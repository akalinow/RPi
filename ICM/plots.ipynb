```python
import json
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

def plot_forecast_data(json_file):
    """
    Plots the forecast data from a JSON file.

    Args:
        json_file (str): The path to the JSON file.
    """
    with open(json_file, 'r') as f:
        data = json.load(f)

    # Assuming only one key-value pair in the loaded json
    key = list(data.keys())[0]
    forecast_data = data[key]["data"]
    forecast_times = data[key]["times"]

    # Convert time strings to datetime objects
    datetime_objects = [datetime.fromisoformat(time.replace("Z", "+00:00")) for time in forecast_times]

    # Create the plot
    plt.figure(figsize=(12, 6))
    plt.plot(datetime_objects, forecast_data, marker='.', linestyle='-')
    plt.xlabel("Time")
    plt.ylabel("Forecast Value")
    plt.title(f"Forecast Data for {key}")
    plt.grid(True)
    plt.xticks(rotation=45)

    # Format x-axis to show dates
    date_format = mdates.DateFormatter('%Y-%m-%d %H:%M')
    plt.gca().xaxis.set_major_formatter(date_format)
    plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    json_file = "forecast_data_05226M0001500.json"
    json_file = "forecast_data_03236_0000000.json"
    plot_forecast_data(temp_json, perp_json)
```
